### 1.1 Preprocessing Steps
During preprocessing, I first loaded the three provided datasets:
final_posts.json, emotion.csv, and data_identification.csv. The JSON file contained the raw Twitter posts in nested structure, so I used pd.json_normalize() to flatten it and extract the fields post_id and text .

Next, I merged the posts with the emotion labels and the train/test identification file using post_id as the key. After merging, I separated the dataset into:

* Training set: rows where emotion is not null

* Test set: rows where emotion is null

To better understand the dataset, I plotted the emotion distribution and text-length histogram. The distribution revealed strong imbalance, with joy and anger being the most frequent classes.

I also implemented a custom text-cleaning function to normalize the posts. The function performed:

* HTML unescaping `(&amp; → &)`

* Replacing URLs with `[URL]`

* Replacing @username mentions with `[USER]`

* Removing the hashtag symbol while keeping the word

* Collapsing repeated whitespace

This cleaned text was stored in a new column text_clean and used consistently for both training and prediction



### 1.2 Feature Engineering Steps

For feature engineering, I focused on transforming the cleaned text into model-ready inputs for a transformer-based classifier. First, I constructed a label mapping (label2id and id2label) for the six emotion categories: anger, disgust, fear, joy, sadness, and surprise.

The cleaned text (text_clean) was used as input to the tokenizer of the model `cardiffnlp/twitter-roberta-base-emotion`. I set max_length=250, which was based on the earlier text-length histogram that showed most posts fall within this range.

I then created a custom PyTorch dataset class EmotionDataset, which tokenized each text into:

* input_ids

* attention_mask

* labels for supervised training

Finally, I split the dataset into train (90%) and validation (10%), using stratified sampling to preserve label distribution. No traditional NLP features (TF–IDF, n-grams, etc.) were added, since transformer models directly learn contextual embeddings from raw text tokens

### 1.3 Explanation of Your Model

The model used for this competition was a fine-tuned transformer:
`cardiffnlp/twitter-roberta-base-emotion`, an emotion-focused RoBERTa variant pretrained specifically on Twitter data. This makes it particularly suitable for short, informal social-media posts.

I loaded the model using `AutoModelForSequenceClassification` with six output labels. Because the checkpoint originally had a different output size, I enabled ignore_mismatched_sizes=True to allow re-initialization of the final classification layer.

For training, I used HuggingFace’s Trainer API with the following hyperparameters:

* Learning rate: 2e-5

* Batch size: 128

* Weight decay: 0.01


Evaluation and checkpoint saving at every epoch

Metric for best model selection: macro F1 (due to label imbalance)

I computed two evaluation metrics: accuracy and macro F1, with macro F1 serving as the primary metric for model selection. After training, I generated predictions for the test set using the cleaned text and saved the results in the required Kaggle submission format.

Finally, I automatically identified the best checkpoint (the latest saved epoch) and copied it to Google Drive for future reuse.

```python
import pandas as pd
import json
import matplotlib.pyplot as plt

posts_path = "/content/final_posts.json"
emotion_path = "/content/emotion.csv"
id_path = "/content/data_identification.csv"

# Load the entire JSON file (it is a list, not a dict)
with open(posts_path, "r", encoding="utf-8") as f:
    data = json.load(f)   # data is a list

# Normalize the list
df_posts = pd.json_normalize(data)

# Extract needed fields
df_posts["post_id"] = df_posts["root._source.post.post_id"]
df_posts["text"] = df_posts["root._source.post.text"]

# Keep only useful columns
posts = df_posts[["post_id", "text"]]

posts.head()
```

```python
# read emotion
emotion = pd.read_csv(emotion_path)

# read train/test split
data_id = pd.read_csv(id_path)

print("=== Loaded Data Shapes ===")
print("posts:", posts.shape)
print("emotion:", emotion.shape)
print("data_identification:", data_id.shape)

# Align key column name with posts (post_id)
emotion = emotion.rename(columns={"id": "post_id"})
data_id = data_id.rename(columns={"id": "post_id"})

# merge emotion label to training set
df = posts.merge(emotion, on="post_id", how="left")
df = df.merge(data_id, on="post_id", how="left")

print("\n=== Merged Data Shape ===")
print(df.shape)

# Training set = data with label
train_df = df[df["emotion"].notnull()]
test_df = df[df["emotion"].isnull()]

print("\nTrain size:", train_df.shape)
print("Test size:", test_df.shape)

# Emotion label
plt.figure(figsize=(8,5))
train_df["emotion"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Emotion Count Distribution")
plt.xlabel("Emotion")
plt.ylabel("Count")
plt.show()

# text length examination
train_df["text_length"] = train_df["text"].apply(lambda x: len(str(x)))

plt.figure(figsize=(8,5))
train_df["text_length"].hist(bins=50)
plt.title("Text Length Distribution")
plt.xlabel("Length")
plt.ylabel("Frequency")
plt.show()

train_df.head()
```

```python
import re
import pandas as pd

def clean_text(text: str) -> str:
    # Handle NaN or non-string input
    if pd.isna(text):
        return ""
    text = str(text)

    # Unescape common HTML entities
    text = text.replace("&amp;", "&")

    # Replace URLs with a special token
    text = re.sub(r"(http\S+|www\.\S+)", " [URL] ", text)

    # Replace @username with a special token
    text = re.sub(r"@\w+", " [USER] ", text)

    # Remove hashtag symbol but keep the word (e.g., #happy -> happy)
    text = re.sub(r"#(\w+)", r"\1", text)

    # Collapse multiple spaces
    text = re.sub(r"\s+", " ", text).strip()

    return text

# Apply to full df so train/test share the same preprocessing
df["text_clean"] = df["text"].apply(clean_text)

# Re-split (optional but clean)
train_df = df[df["emotion"].notnull()].copy()
test_df = df[df["emotion"].isnull()].copy()

# Quick check
train_df[["text", "text_clean"]].head()
```

```python
import numpy as np
from sklearn.model_selection import train_test_split

# Make sure we only use rows with non-null label
train_df = df[df["emotion"].notnull()].copy()
test_df = df[df["emotion"].isnull()].copy()

# Use cleaned text for BERT input
train_texts = train_df["text_clean"].tolist()
train_labels_str = train_df["emotion"].tolist()

# Build label mapping
unique_labels = sorted(train_df["emotion"].unique())
label2id = {label: idx for idx, label in enumerate(unique_labels)}
id2label = {idx: label for label, idx in label2id.items()}

num_labels = len(unique_labels)
print("Label mapping:", label2id)

# Convert string labels to integer ids
train_labels = [label2id[l] for l in train_labels_str]

# Train/validation split
X_train, X_valid, y_train, y_valid = train_test_split(
    train_texts,
    train_labels,
    test_size=0.1,
    random_state=42,
    stratify=train_labels
)

len(X_train), len(X_valid)
```

```python
from transformers import BertTokenizer
from transformers import AutoTokenizer
tokenizer = AutoTokenizer.from_pretrained("cardiffnlp/twitter-roberta-base-emotion")

import torch
model_name = "cardiffnlp/twitter-roberta-base-emotion"

max_length = 250  # based on text_length distribution

class EmotionDataset(torch.utils.data.Dataset):
    def __init__(self, texts, labels=None, tokenizer=None, max_length=128):
        self.texts = texts
        self.labels = labels
        self.tokenizer = tokenizer
        self.max_length = max_length

    def __len__(self):
        return len(self.texts)

    def __getitem__(self, idx):
        text = str(self.texts[idx])

        encodings = self.tokenizer(
            text,
            truncation=True,
            padding="max_length",
            max_length=self.max_length,
            return_tensors="pt"
        )

        item = {
            "input_ids": encodings["input_ids"].squeeze(0),
            "attention_mask": encodings["attention_mask"].squeeze(0),
        }

        if self.labels is not None:
            item["labels"] = torch.tensor(self.labels[idx], dtype=torch.long)

        return item

import torch

train_dataset = EmotionDataset(X_train, y_train, tokenizer, max_length)
valid_dataset = EmotionDataset(X_valid, y_valid, tokenizer, max_length)
```

```python
from transformers import BertForSequenceClassification, TrainingArguments, Trainer
import evaluate
from transformers import AutoModelForSequenceClassification
model = AutoModelForSequenceClassification.from_pretrained(
    "cardiffnlp/twitter-roberta-base-emotion",
    num_labels=num_labels,
    id2label=id2label,
    label2id=label2id,
    ignore_mismatched_sizes=True
)

# Metric: accuracy + macro F1 (for imbalanced emotions)
accuracy = evaluate.load("accuracy")
f1 = evaluate.load("f1")

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)

    acc = accuracy.compute(predictions=preds, references=labels)["accuracy"]
    f1_macro = f1.compute(predictions=preds, references=labels, average="macro")["f1"]

    return {"accuracy": acc, "f1_macro": f1_macro}

batch_size = 128
epochs = 3

training_args = TrainingArguments(
    output_dir="/content/bert-emotion-checkpoints",
    eval_strategy="epoch",
    save_strategy="epoch",
    report_to="none",
    learning_rate=2e-5,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    num_train_epochs=epochs,
    weight_decay=0.01,
    load_best_model_at_end=True,
    metric_for_best_model="f1_macro",
    logging_steps=100,
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=valid_dataset,
    tokenizer=tokenizer,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
trainer.evaluate()
```

```python
# Build test dataset using text_clean
test_texts = test_df["text_clean"].tolist()
test_dataset = EmotionDataset(test_texts, labels=None, tokenizer=tokenizer, max_length=max_length)

# Get predictions
predictions = trainer.predict(test_dataset)
logits = predictions.predictions
test_preds_ids = np.argmax(logits, axis=-1)

# Map back to label names
test_preds_labels = [id2label[i] for i in test_preds_ids]

# Attach to test_df
test_df["emotion_pred"] = test_preds_labels
test_df[["post_id", "emotion_pred"]].head()
```
